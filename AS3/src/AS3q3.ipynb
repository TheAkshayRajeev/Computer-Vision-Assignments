```python
import cv2
import numpy as np
import math
from keras.models import load_model
```

```python
def input_image():
    file = input("Enter the path:")
    image_original = cv2.imread(file,1)
    return image_original
```

```python
def standardazie(image):
    image_mean = np.mean(image)
    image_std  = np.std(image)
    standardized_image = (image - image_mean) / image_std
    return standardized_image
```

```python
def pre_processing(image):
    grey_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    resized_image = cv2.resize(grey_image, (28,28), interpolation = cv2.INTER_AREA)
    reshaped_image = resized_image.reshape(28, 28, 1)
    ret,threshold = cv2.threshold(reshaped_image,127,255,cv2.THRESH_BINARY)
    threshold = threshold - 255
    standardized_threshold = standardazie(threshold)
    blur_image = cv2.GaussianBlur(standardized_threshold,(5,5),0)
    preprocessed_image = blur_image.reshape(1,28,28,1)
    return preprocessed_image
```
